In this activity, we will use the NumPy to create a tic-tac-toe (noughts and crosses) simulator and evaluate basic winning strategies.

```python
import numpy as np
import random 

def create_board():
    return (np.zeros(shape=(3,3))).astype(int)
def place(board, player:int, position:list):
    board[position] = player

def possibilities(board):
    indexes = []
    for rows in range(board.shape[0]):
        for cols in range(board.shape[1]):
            if(board[rows,cols] == 0):
                indexes.append((rows,cols))
    return indexes

def random_place(board, player, available):
    place = random.choice(available)
    # print("fff")
    board[place] = player
    return

def row_win(board, player):
    flag = True
    for rows in board:
        flag = True
        for cols in rows:
            if(cols != player):
                flag = False
        if (flag == True):
            # print(rows)
            return True
    return False

def col_win(board, player):
    flag = False
    for rows in range(board.shape[1]):
        if(board[0][rows] == player and board[1][rows]== player and board[2][rows] == player):
            flag = True
    return flag

def diag_win(board, player):
    flag = True
    size = board.shape[0] - 1
    
    j = 0
    for i in range(size+1):
        # print(board[i][i])
        if(board[i][i] != player):
            flag = False
    if(flag == True):
        return flag
    flag = True
    for i in range(size,-1,-1):
        # print(j,i)
        if(board[j][i] != player):
            flag = False
        j += 1

    return flag


```

```python
# random.seed(1)

def random_evaluate(board):
    winner = 0
    player = 1

    while(winner == 0):

        if(np.all(board != 0)):
            winner = -1

        available = possibilities(board)
        if(available):
            random_place(board,player,available)
        if(row_win(board,player)):
            # print("row")
            winner = player
        elif(col_win(board,player)):
            # print("col")
            winner = player
        elif(diag_win(board,player)):
            # print("diag")
            winner = player
        else:
            if(player == 1):
                player = 2
            else:
                player = 1
        
    if winner == 0:
        winner = -1
    return winner

def strategy_one(board):
    winner = 0
    player = 2

    place(board,1,(1,1))
    # print(board)

    while(winner == 0):

        if(np.all(board != 0)):
            winner = -1

        available = possibilities(board)
        if(available):
            random_place(board,player,available)
        if(row_win(board,player)):
            # print("row")
            winner = player
        elif(col_win(board,player)):
            # print("col")
            winner = player
        elif(diag_win(board,player)):
            # print("diag")
            winner = player
        else:
            if(player == 1):
                player = 2
            else:
                player = 1
        
    if winner == 0:
        winner = -1
    return winner


def play_game():
    board = create_board()
    result = strategy_one(board)
    return result


def play_random_game():
    board = create_board()
    result = random_evaluate(board)
    return result

winner_record = []

for i in range(1000):
    winner_record.append(play_random_game())

print(winner_record.count(1))
print(winner_record.count(-1))
print(winner_record.count(2))

winner_record = []
print()
for i in range(1000):
    winner_record.append(play_game())

print(winner_record.count(1))
print(winner_record.count(-1))
print(winner_record.count(2))


```
